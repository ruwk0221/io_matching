```python
import sys
import os
sys.path.append(os.path.dirname(os.path.abspath(os.path.dirname('__file__'))))
import jReversion as jR
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score, recall_score, precision_score, f1_score, balanced_accuracy_score
import networkx as nx
from statannot import add_stat_annotation
from lifelines import KaplanMeierFitter
from lifelines.statistics import logrank_test
kmf = KaplanMeierFitter()
```

```python
oncoprint_pd = pd.read_csv('../data/BLCA_200919/CCLE_DATA_oncoprint.tsv', sep='\t')
print(oncoprint_pd)
```

```python
oncoprint_pd = pd.melt(oncoprint_pd, id_vars=['track_name', 'track_type'], var_name='Case ID', value_name='event')
print(oncoprint_pd)
```

```python
oncoprint_blca_pd = oncoprint_pd[oncoprint_pd['Case ID'].str.contains('URINAR')]
oncoprint_blca_pd['cell line'] = oncoprint_blca_pd['Case ID'].str.replace('_URINARY_TRACT', '')
print(oncoprint_blca_pd.head())
```

```python
gene_node_pd = pd.read_csv('../data/BLCA_200919/Grieco_gene_node_mapping.csv')
# print(gene_node_pd.head())
table_original_network = pd.read_csv('../data/Grieco_table_for_original_network.tsv', sep='\t').sort_values(by='node')
```

```python
event_count_pd = pd.merge(oncoprint_pd, gene_node_pd, left_on='track_name', right_on='gene_symbol').set_index(['track_name', 'track_type', 'node_name']).count(level='node_name')
# event_count_pd.sort_values('event', ascending=False).head(20)
```

```python
result = pd.merge(event_count_pd, table_original_network, left_on='node_name', right_on='node')
fig, ax = plt.subplots(1, 1, figsize=(10,10))
fig.suptitle('BLCA CCLE mutation count w/ passenger', fontsize=15, y=0.93)
ax = sns.stripplot(data=result.sort_values('Class'), x='Class', y='event')
ax = sns.boxplot(data=result.sort_values('Class'), x='Class', y='event', color='.9', ax=ax)

add_stat_annotation(ax=ax, data=result.sort_values('Class'), x='Class', y='event',
                    box_pairs=[("C0", "C1"), ("C1", "C2"), ("C1", "C3"), ("C0", "C3"), ("C2", "C3"), ("C0", "C2")],
                    test='Mann-Whitney', text_format='star', loc='inside', verbose=2)
# plt.savefig('../data/BLCA_200919/alter_freq_BLCA.pdf', dpi=300)
```

```python
oncoprint_node_pd = pd.merge(oncoprint_blca_pd, gene_node_pd, left_on='track_name', right_on='gene_symbol')
oncoprint_node_class_pd = pd.merge(oncoprint_node_pd.dropna(), table_original_network[['node', 'Class']], left_on='node_name', right_on='node')
oncoprint_node_class_pd['alteration'] = 1
print(oncoprint_node_class_pd.head())
oncoprint_reverse_controllability = pd.pivot_table(oncoprint_node_class_pd[['cell line', 'Class', 'alteration']], index='cell line', columns='Class', values='alteration', aggfunc=np.sum)
oncoprint_reverse_controllability = oncoprint_reverse_controllability.fillna(0)
oncoprint_reverse_controllability['score'] = oncoprint_reverse_controllability['C2'] + oncoprint_reverse_controllability['C3'] - oncoprint_reverse_controllability['C1']
print(oncoprint_reverse_controllability.head())
```

```python
sns.countplot(data=oncoprint_reverse_controllability, x='score')
```

```python
pharmaco_profile = pd.read_csv('../bladder_cell_line/BLCA_IC_Tue Jul  7 09_26_30 2020.csv')
pharmaco_profile['cell line'] = pharmaco_profile['Cell line name'].str.replace('-', '')
print(pharmaco_profile.head())
pivot_pharmaco_profile_ic50 = pharmaco_profile.pivot(index='cell line', columns='Drug Id', values='IC50')
#pivot_pharmaco_profile_auc = pharmaco_profile.pivot(index='Cell line name', columns='Drug Id', values='AUC')
pivot_pharmaco_profile_zscore = pharmaco_profile.pivot(index='cell line', columns='Drug Id', values='Z score')
drug_info = pd.read_excel('../bladder_cell_line/gdsc_drugs.xls', sheet_name='Sheet1')
drug_info = drug_info.set_index('drug_id')
# cell_line_mapping = oncoprint_reverse_controllability
# cell_line_set = cell_line_mapping.index
# print(cell_line_mapping.head())
```

```python
tmp_ic50 = pd.merge(oncoprint_reverse_controllability, pivot_pharmaco_profile_ic50, left_index=True, right_index=True, how='inner')
#tmp_auc = pd.merge(reverse_controllability_cell_line, pivot_pharmaco_profile_auc, left_index=True, right_index=True, how='left')
tmp_zscore = pd.merge(oncoprint_reverse_controllability, pivot_pharmaco_profile_zscore, left_index=True, right_index=True, how='inner')
# print(tmp.head())
pearson_ic50 = tmp_ic50.corr()
pearson_ic50.drop(['C0', 'C1', 'C2', 'C3', 'score'], inplace=True)
pearson_ic50 = pearson_ic50.filter(items=['C0', 'C1', 'C2', 'C3', 'score'])

pearson_zscore = tmp_zscore.corr()
pearson_zscore.drop(['C0', 'C1', 'C2', 'C3', 'score'], inplace=True)
pearson_zscore = pearson_zscore.filter(items=['C0', 'C1', 'C2', 'C3', 'score'])

#print(pearson.head())
from scipy.stats import kendalltau, pearsonr, spearmanr

def pearsonr_pval(x,y):
    return pearsonr(x,y)[1]


pearson_p_ic50 = tmp_ic50.corr(method=pearsonr_pval)
pearson_p_ic50.drop(['C0', 'C1', 'C2', 'C3', 'score'], inplace=True)
pearson_p_ic50 = pearson_p_ic50.filter(items=['C0', 'C1', 'C2', 'C3', 'score'])
pearson_p_zscore = tmp_zscore.corr(method=pearsonr_pval)
pearson_p_zscore.drop(['C0', 'C1', 'C2', 'C3', 'score'], inplace=True)
pearson_p_zscore = pearson_p_zscore.filter(items=['C0', 'C1', 'C2', 'C3', 'score'])

pearson_ic50 = pd.merge(pearson_ic50, pearson_p_ic50,left_index=True, right_index=True, how='left', suffixes=('_corr', '_pval'))
pearson_ic50 = pd.merge(pearson_ic50, drug_info,left_index=True, right_index=True, how='left')

pearson_zscore = pd.merge(pearson_zscore, pearson_p_zscore,left_index=True, right_index=True, how='left', suffixes=('_corr', '_pval'))
pearson_zscore = pd.merge(pearson_zscore, drug_info,left_index=True, right_index=True, how='left')

col_names = [str(x) for x in tmp_ic50.columns]
tmp_ic50.columns = col_names

```

```python
print(tmp_ic50.head())
print(pearson_ic50.head())
```

```python
score = 'score'
meaningful = pearson_ic50[score+'_pval'] < 0.05
r, c = pearson_ic50.loc[meaningful].shape
drug_ids = [str(x) for x in pearson_ic50.loc[meaningful].index]
print(drug_ids)
print(r)
number_r = int(np.ceil(r / 3))
print(number_r)

def label_point(df,drug_id, ax):
    for i, point in df.iterrows():
        ax.annotate(i, xy = (point[score], point[str(drug_id)]),
                    xytext=(2,2), textcoords="offset points")

fig, axes = plt.subplots(number_r,3, sharex=True, sharey=True, figsize=(15,number_r * 6))
#drug_ids = ['1916', '1924', '2170', '1940']
for idx, ax in enumerate(axes.flat):        
    #print(ax)
    if idx < len(drug_ids):
        sns.regplot(x=tmp_ic50[score], y=tmp_ic50[drug_ids[idx]], ax=ax)
        label_point(tmp_ic50, drug_ids[idx], ax)
    else:
        break
```
