```python
import pandas as pd
from matplotlib import pyplot as plt
from lifelines import KaplanMeierFitter
from lifelines.statistics import logrank_test
kmf = KaplanMeierFitter()
```

```python
c0_data = pd.read_csv('../data/Overall_BLCA_TCGA_Cell_2017_C0.csv')
c1_data = pd.read_csv('../data/Overall_BLCA_TCGA_Cell_2017_C1.csv')
c2_data = pd.read_csv('../data/Overall_BLCA_TCGA_Cell_2017_C2.csv')
c3_data = pd.read_csv('../data/Overall_BLCA_TCGA_Cell_2017_C3.csv')

f, axes = plt.subplots(2, 2, figsize=(10, 10), sharex=True, sharey=True)
f.suptitle('BLCA Overall survival', fontsize=15, y=0.93)

ax = axes[0,0]
data = c0_data
T = data['Time (months)']
E = data['Status'] == 'deceased'
altered = (data['Group'] == 'altered')
kmf.fit(T[altered], event_observed=E[altered], label = 'C0 Altered')
kmf.plot(ax=ax)
kmf.fit(T[~altered], event_observed=E[~altered], label = 'C0 Unaltered')
kmf.plot(ax=ax)
results = logrank_test(T[altered], T[~altered], E[altered], E[~altered], alpha=.99)
ax.text(0.1, 0.1, 'p-value=%.3f'% results.p_value)
ax.set_title('C0')
ax.set_xlabel('Time (monts)')
ax.set_ylabel('Overall survival')

ax = axes[0,1]
data = c1_data
T = data['Time (months)']
E = data['Status'] == 'deceased'
altered = (data['Group'] == 'altered')
kmf.fit(T[altered], event_observed=E[altered], label = 'C1 Altered')
kmf.plot(ax=ax)
kmf.fit(T[~altered], event_observed=E[~altered], label = 'C1 Unaltered')
kmf.plot(ax=ax)
results = logrank_test(T[altered], T[~altered], E[altered], E[~altered], alpha=.99)
ax.text(0.1, 0.1, 'p-value=%.3f'% results.p_value)
ax.set_title('C1')
ax.set_xlabel('Time (monts)')
ax.set_ylabel('Overall survival')

ax = axes[1,0]
data = c2_data
T = data['Time (months)']
E = data['Status'] == 'deceased'
altered = (data['Group'] == 'altered')
kmf.fit(T[altered], event_observed=E[altered], label = 'C2 Altered')
kmf.plot(ax=ax)
kmf.fit(T[~altered], event_observed=E[~altered], label = 'C2 Unaltered')
kmf.plot(ax=ax)
results = logrank_test(T[altered], T[~altered], E[altered], E[~altered], alpha=.99)
ax.text(0.1, 0.1, 'p-value=%.3f'% results.p_value)
ax.set_title('C2')
ax.set_xlabel('Time (monts)')
ax.set_ylabel('Overall survival')

ax = axes[1,1]
data = c3_data
T = data['Time (months)']
E = data['Status'] == 'deceased'
altered = (data['Group'] == 'altered')
kmf.fit(T[altered], event_observed=E[altered], label = 'C3 Altered')
kmf.plot(ax=ax)
kmf.fit(T[~altered], event_observed=E[~altered], label = 'C3 Unaltered')
kmf.plot(ax=ax)
results = logrank_test(T[altered], T[~altered], E[altered], E[~altered], alpha=.99)
ax.text(0.1, 0.1, 'p-value=%.3f'% results.p_value)
ax.set_title('C3')
ax.set_xlabel('Time (monts)')
ax.set_ylabel('Overall survival')
```
