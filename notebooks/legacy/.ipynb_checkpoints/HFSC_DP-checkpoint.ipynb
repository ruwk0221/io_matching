```python
import sys
import os
sys.path.append(os.path.dirname(os.path.abspath(os.path.dirname('__file__'))))
import jReversion as jR
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score, recall_score, precision_score, f1_score, balanced_accuracy_score
import networkx as nx
from statannot import add_stat_annotation
import robustness as rb
import concurrent.futures
import par_helper as ph
```

```python
model = jR.initialize('../models/HFSC.txt', 'n', 'n')

mapping = model['mapping']
inverseMapping = model['inverse_mapping']
gRead = model['Gread']
readNodes = model['read_nodes']
# if model['reduction_required']:
#     gRead = BDOIp.Get_reduced_network(gRead)

# gExp = ph.par_get_expanded_network(gRead, prefix='n', suffix='n', worker=5)

bit_num = len(gRead)
state_num = 2**15
initial_states = rb.gen_rand_bin_state(bit_num, state_num)
wt_result = rb.identifying_synchronous_attractors_basins(gRead, initial_states.copy())

renamed_gRead = nx.DiGraph()
renamed_gRead.add_edges_from(gRead.edges())
renamed_gRead = nx.relabel_nodes(renamed_gRead, readNodes)
# nx.write_gml(renamed_gRead, path='../networks/HFSC_DP.gml')

inverseReadNodes = dict()
for key, val in readNodes.items():
    inverseReadNodes[val] = key

node_activity_df = pd.DataFrame()
for attractor in wt_result['attractors']:
    activity_dict = dict()
    if wt_result['type_of_attractor'][attractor] == 'point':
        activity_dict = {readNodes[idx]: float(val) for idx, val in enumerate(attractor)}
    else:
        temp = []
        for state in wt_result['states_in_attractor'][attractor]:
            temp.append([int(x) for x in state])
        activity_dict = {readNodes[idx]: float(val) for idx, val in enumerate(np.array(temp).mean(axis=0))}
        
    node_activity_df = node_activity_df.append(activity_dict, ignore_index=True)


# node_activity_df.to_csv('../data/HFSC_DP_node_activity.csv')
```

```python
g = sns.clustermap(node_activity_df, cmap="vlag")
```

```python
# TODO 
# basin size  CDK4 activity   basin size  mapping
sampled_space = sum(wt_result['basin_of_attractor'].values())
node_score_df = pd.DataFrame()
for attractor in wt_result['attractors']:
    score_dict = dict()
    basin_ratio = wt_result['basin_of_attractor'][attractor] / float(sampled_space)
    if wt_result['type_of_attractor'][attractor] == 'point':
        score_dict = {readNodes[idx]: float(val)for idx, val in enumerate(attractor)}
    else:
        temp = []
        for state in wt_result['states_in_attractor'][attractor]:
            temp.append([int(x) for x in state])
        score_dict = {readNodes[idx]: float(val) for idx, val in enumerate(np.array(temp).mean(axis=0))}
    score_dict['basin_ratio'] = basin_ratio
    node_score_df = node_score_df.append(score_dict, ignore_index=True)
```
