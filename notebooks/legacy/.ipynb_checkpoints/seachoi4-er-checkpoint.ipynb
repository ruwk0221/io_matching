```python
def fvs_max_ldoi_condition(G, mapping, inverse_mapping, subnetwork, fvs, input_condition):
#     fvs = ['AKT1', 'CEP55', 'CCNE1', 'CCND1', 'ZEB1', 'SNAI1', 'PI3K']
    fvs_conditions = np.ndarray((len(fvs), 2), dtype=object)
    for idx, fvs_node in enumerate(fvs):
        fvs_conditions[idx, 0] = '~' + fvs_node
        fvs_conditions[idx, 1] = fvs_node

    stabled_nodes_list = dict()    

    for i in range(2 ** len(fvs)):
    #     G_copied = G_expanded.copy()
        temp = np.binary_repr(i, width=len(fvs))
        fvs_condition = [fvs_conditions[x, int(temp[x])] for x in range(len(fvs))]

        intervention = [mapping[x] for x in fvs_condition] + [mapping[x] for x in input_condition]
        LDOI_raw, _, _, _ = BDOId.truncated_node_of_influence_BFS(G, intervention)
        LDOI = set([x for x in LDOI_raw if x.find('_') < 0])
        LDOI_gene = [inverse_mapping[x] for x in LDOI]
    #     ldoi_fvs_conditions[temp] = LDOI_gene.copy()

        gene = [x.replace('~', '') for x in LDOI_gene]
        ldoi_and_fvs = set(gene + [x.replace('~', '') for x in fvs])
        stabled_nodes_list[temp] = set(subnetwork).intersection(ldoi_and_fvs)

    max_condition = ''
    max_stabled_num_nodes = 0
    for key, val in stabled_nodes_list.items():
        if len(val) > max_stabled_num_nodes:
            max_condition = key
            max_stabled_num_nodes = len(val)

#     print(max_condition + ': ' + str(max_stabled_num_nodes))

    max_condition_list = []
    for key, val in stabled_nodes_list.items():
        if len(val) == max_stabled_num_nodes:
            max_condition_list.append(key)

    stable_count = dict()
    for node in subnetwork:
        stable_count[node] = 0

    for condition in max_condition_list:
        for node in stabled_nodes_list[condition]:
            stable_count[node] += 1
            
    return {
            'max_condition': max_condition,
            'max_stabled_num_nodes': max_stabled_num_nodes,
            'max_condiiton_list': max_condition_list,
            'stable_count': stable_count
    }
```

```python
def fvs_control(G_ex, mapping, inverse_mapping, subnetwork, pd, fvs, threshold=0):
    intervention = []
    for node in fvs:
        if (pd[pd['node'] == node]['diff_abs'] > threshold).all():
            if (pd[pd['node'] == node]['average difference']< 0).all():
                intervention.append('~'+node)
            else:
                intervention.append(node)
    #     print(intervention)    
    temp = dict()
    temp['intervention'] = intervention.copy()
    LDOI_raw, _, _, _ = BDOId.truncated_node_of_influence_BFS(G_ex, [mapping[x] for x in intervention])
    LDOI = set([x for x in LDOI_raw if x.find('_') < 0])
    LDOI_gene = [inverse_mapping[x] for x in LDOI]
    #     ldoi_fvs_conditions[temp] = LDOI_gene.copy()
    temp['ldoi'] = LDOI_gene.copy()
    gene = [x.replace('~', '') for x in LDOI_gene]
    ldoi_and_fvs = set(gene + [x.replace('~', '') for x in fvs])
    stabled_nodes_list = set(subnetwork).intersection(ldoi_and_fvs)
#     temp['ldoi'] = LDOI_gene.copy()
    temp['stable_nodes'] = stabled_nodes_list.copy()
    temp['stabled_nodes_count'] = len(stabled_nodes_list)
    temp['controlled_nodes_count'] = len(set(subnetwork).intersection(set(gene)))
    
    return temp.copy()
```

```python
fvs_scc_65 = [['PI3K', 'NCOR1', 'CENPF', 'STK11', 'CCND1'], ['PI3K', 'NCOR1', 'CEP55', 'STK11', 'CCND1']]
results_scc_65 = dict()
# for idx, fvs in enumerate(fvs_scc_65):
#     temp = fvs_max_ldoi_condition(G=breast_scc_65_ex, mapping=Mapping, inverse_mapping=InverseMapping, subnetwork=scc_65, fvs=fvs, input_condition=[])
#     temp['fvs'] = str(fvs)
#     results_scc_65[idx] = temp.copy()
    
for idx, fvs in enumerate(fvs_scc_65):
    results_scc_65[idx] = fvs_control(G_ex=breast_scc_65_ex, mapping=Mapping, inverse_mapping=InverseMapping, subnetwork=scc_65, pd=erbb_pd, fvs=fvs, threshold=0.05)
```

```python
def fvs_control_1_del(G_ex, mapping, inverse_mapping, subnetwork, pd, fvs, controlled_nodes_count, threshold=0):
    result_1_del = dict()
    for node in fvs:
        fvs_temp = fvs.copy()
        fvs_temp.remove(node)
        temp = fvs_control(G_ex=G_ex, mapping=mapping, inverse_mapping=InverseMapping, subnetwork=subnetwork, pd=erbb_pd, fvs=fvs_temp, threshold=threshold)
        if temp['controlled_nodes_count'] == controlled_nodes_count:
            temp['is_critical'] = False
        else:
            temp['is_critical'] = True
        result_1_del[node] = temp.copy()
    return result_1_del
```
